```python
import os,numpy as np,pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
path=os.getcwd()+"/../Models/EmbryonClassif/embryon_classif/"
print(path)
f=os.listdir(path)
print(f[0])
```

```python
df=pd.read_csv(path+f[0]+"/results.csv")
print(len(df))
cm=confusion_matrix(df['Class'], df['preds'],labels=[0,1,2,3,4,5,6,7])
m_display = ConfusionMatrixDisplay(cm).plot()
```
