```python
import os
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
path =  os.getcwd() +"/../../"  
pathrawdata = path +"/data/rawdata/"
pathsegmented= path +"/results/segmented_videos/"
#print(os.listdir(path),os.listdir(pathrawdata),os.listdir(pathsegmented))

file_name="V83.mp4"
cap = cv2.VideoCapture(pathrawdata+file_name)
cap1 = cv2.VideoCapture(pathsegmented+file_name)
print(file_name)

w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

frame_number = 6
fig, axs = plt.subplots(1,frame_number,figsize=(18, 4))
fig.suptitle(file_name+" RawData")

figure,ax = plt.subplots(1,frame_number,figsize=(18, 4))
figure.suptitle(file_name+" SegmentedData")

count=0
k=0

while(cap.isOpened() and cap1.isOpened()):
    ret, frame = cap.read()
    ret1 , frame1 = cap1.read()
    if ret==True and ret1 == True :
        count+=1
        if count % (300/frame_number) == 0:
            axs[k].imshow(frame)
            axs[k].set_title(str(count))
            ax[k].imshow(frame1)
            ax[k].set_title(str(count))
            k+=1
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
count=0
k=0
cap1.release
cap.release()
fig.tight_layout()
figure.tight_layout()
plt.show()
cv2.destroyAllWindows()
```
